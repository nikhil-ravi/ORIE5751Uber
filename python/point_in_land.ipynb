```python
import pandas as pd
import h3
```

```python
df = pd.read_csv('../datasets/uber_super_cleaned.csv')
```

```python
df.describe()
```

```python
h3_dist = []
for i in range(len(df)):
    h3_dist.append(h3.point_dist((df.pick_lat[i],df.pick_lon[i]),(df.drop_lat[i],df.drop_lon[i]), unit='m'))
df['h3_dist'] = h3_dist
```

```python
df.to_csv('../datasets/uber_super_cleaned.csv')
```
